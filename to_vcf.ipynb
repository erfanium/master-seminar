```python
import os
import re
import pandas as pd

outdir = "./data-v3/00_raw_vcf"
os.makedirs(outdir, exist_ok=True)

zygo_to_gt = {
    "Het": "0/1",
    "Hom": "1/1",
    "Hemi": "1",
    "Hem": "1",
}

def chrom_key(c):
    """
    Enforce VCF chromosome ordering:
    1–22, X, Y, MT, then anything else
    """
    c = str(c).replace("chr", "")
    if c.isdigit():
        return (0, int(c))
    if c == "X":
        return (1, 23)
    if c == "Y":
        return (1, 24)
    if c in {"M", "MT"}:
        return (1, 25)
    return (2, c)


def normalize_ad(ad):
    """
    Ensure AD is exactly two values: ref,alt
    Return '.,.' if invalid.
    """
    if pd.isna(ad):
        return ".,."
    ad = str(ad).replace("-", ",")
    parts = ad.split(",")
    if len(parts) != 2:
        return ".,."
    if not all(p.isdigit() for p in parts):
        return ".,."
    return ad


for i, (pid, group) in enumerate(csv.groupby("Patient_ID")):
    # # stop after i > 5
    # if i > 2:
    #     break
    
    safe_pid = re.sub(r"[^A-Za-z0-9_.-]", "_", pid)
    path = os.path.join(outdir, f"{safe_pid}.vcf")

    # split Variant -> CHROM, POS, REF, ALT
    # expects format like "19-58858382-G-A"
    parts = group["Variant"].str.replace('--', '-.').str.split("-", n=3, expand=True)

    group = group.assign(
        CHROM=parts[0].astype(str),
        POS=parts[1].astype(int),
        REF=parts[2].astype(str),
        ALT=parts[3].astype(str),
    )

    # sort by CHROM, POS (VCF-compliant)
    group = group.assign(
        _chrom_key=group["CHROM"].map(chrom_key)
    ).sort_values(
        by=["_chrom_key", "POS"]
    ).drop(columns="_chrom_key")

    chrom = group["CHROM"].astype(str)
    pos = group["POS"].astype(str)
    ref = group["REF"]
    alt = group["ALT"]

    qual = group["Quality"].astype(str).fillna(".")
    flt = group["Filter"].astype(str).fillna(".")
    info = "DP=" + group["Depth"].astype(str) + ";ZYG=" + group["Zygosity"].astype(str)
    ad = group["AD"].apply(normalize_ad)
    gt = group["Zygosity"].map(zygo_to_gt).fillna("./.")
    sample_field = gt + ":" + ad + ":" + group["Depth"].astype(str)

    header_lines = [
        f"#CHROM\tPOS\tID\tREF\tALT\tQUAL\tFILTER\tINFO\tFORMAT\t{pid}\n",
    ]

    records = (
        chrom + "\t"
        + pos + "\t.\t"
        + ref + "\t"
        + alt + "\t"
        + qual + "\t"
        + flt + "\t"
        + info + "\tGT:AD:DP\t"
        + sample_field + "\n"
    ).tolist()

    with open(path, "w") as fh:
        fh.writelines([header_common])
        fh.writelines(header_lines)
        fh.writelines(records)

    if (i + 1) % 1000 == 0:
        print(f"Written {i + 1} patient VCFs")
```
